# Comparing SuSie-inf with SusieR package

The base SuSiE package on R vs Susie-inf from Finucanelab https://github.com/FinucaneLab/fine-mapping-inf

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

from gentropy.method.susie_inf import SUSIE_inf
```

```python
z = pd.read_csv("../z_500.csv", index_col=0)
z = np.array(z.iloc[:, 0])
ld = pd.read_csv("../ld_500.csv", index_col=0)
ld = np.array(ld)
susie_inf = SUSIE_inf.susie_inf(z=z, LD = ld)
```

```python
first_lbf_susie_inf = susie_inf["lbf_variable"][:, 0]
second_lbf_susie_inf = susie_inf["lbf_variable"][:, 1]
first_lbf_susie_r = np.array(pd.read_csv("../scripts/first_lbf_susie_r.csv", index_col=0))
first_lbf_susie_r = np.reshape(first_lbf_susie_r, (500))
second_lbf_susie_r = np.array(pd.read_csv("../scripts/second_lbf_susie_r.csv", index_col=0))
second_lbf_susie_r = np.reshape(second_lbf_susie_r, (500))
```

```python
# plt.scatter(x=first_lbf_susie_r, y=first_lbf_susie_inf)
# plt.xlabel("susie_r")
# plt.ylabel("susie_inf")
# plt.show()

corr = pearsonr(first_lbf_susie_r, first_lbf_susie_inf)
print("Pearson's correlation coefficient: " + str(corr[0]))  # noqa: T201
```

```python
# plt.scatter(x=second_lbf_susie_r, y=second_lbf_susie_inf)
# plt.xlabel("susie_r")
# plt.ylabel("susie_inf")
# plt.show()

corr = pearsonr(second_lbf_susie_r, second_lbf_susie_inf)
print("Pearson's correlation coefficient: " + str(corr[0]))  # noqa: T201
```
